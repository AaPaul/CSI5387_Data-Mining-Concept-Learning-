```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

## Data Loading and Preprocessing

```python
with open("C:\\Users\\admin\\Desktop\\work\\assignment\\Dataset 3\\groceries_new.csv", 'r+') as f:
    content = f.read()
    f.seek(0,0)
    str1 = ""
    for i in range(32):
        if (i<31):
            str1 = str1+str(i)+','
        else: 
            str1 = str1+str(i)+'\n'
    f.write(str1+content)
```

```python
df1 = pd.read_csv(".\\Dataset 3\\groceries_new.csv")
df1 = df1.fillna(0)
test1 = df1.values.tolist()
```

```python
for i in range(len(test1)):
    test1[i] = list(set(test1[i]))
    for j in test1[i]:
        if (j):
            continue
        else:
            test1[i].remove(j)
```

```python
def newApriori(df, min_sup, min_conf):
    '''
    df: dataset
    min_sup: minimum support
    min_conf: minimum confidence
    '''
    te = TransactionEncoder()
    te_ary = te.fit(df).transform(df)
    df = pd.DataFrame(te_ary, columns=te.columns_)
    fi = apriori(df, min_support=min_sup, use_colnames=True, max_len=3,verbose=0)
#     print(fi.head())
    display(fi.head())
    rule = association_rules(fi, metric="confidence", min_threshold=min_conf)
#     rule.sort_values('lift',ascending=False).head(5)
#     print(rule.sort_values('lift',ascending=False).head(5))
    return rule
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" 
r1 = newApriori(test1, 0.005, 0.5)
r1.sort_values('lift',ascending=False).head()
```

```python
r2 = newApriori(test1, 0.001, 0.5)
r2.sort_values('lift',ascending=False).head()
```
