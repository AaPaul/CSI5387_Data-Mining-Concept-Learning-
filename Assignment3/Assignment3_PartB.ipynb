```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### Get the initial centroid for cluseter_A & cluseter_B

```python
cluster_A = [[1, 0], [2, 1]]
cluster_B = [[0, 1], [3, 3]]

def SSE(a, b):
    A = np.array(a)
    B = np.array(b)
    A_Mean, B_Mean = cal_Mean(A, B)
    e1 = e2 = 0
    for i in A:
        e1 += np.sum((i-A_Mean)**2)
    for i in B:
        e2 += np.sum((i - B_Mean)**2)
    sse = e1 + e2
    return sse

def cal_Mean(a, b):
    s_a = s_b = 0
    for i in a:
        s_a += i
    aMean = s_a / len(a)
    for j in b:
        s_b += j
    bMean = s_b / len(b)
    return aMean, bMean

print(SSE(cluster_A, cluster_B))
```

```python
X = [[1, 0], [0, 1], [2, 1], [3, 3]]
```

#### Create K-Means model

```python
km = KMeans(n_clusters=2, max_iter=1)
```

#### The first iteration

```python
km.fit(X)
t1 = km.predict(X)
t1
```

```python
c1 = [[1, 0], [0, 1], [2, 1]]
c2 = [[3, 3]]
print(SSE(c1, c2))
```

#### The second iteration

```python
km.fit(X)
t2 = km.predict(X)
t2
```

```python
c1 = [[3, 3]]
c2 = [[1, 0], [0, 1], [2, 1]]
print(SSE(c1, c2))
```

```python
df = pd.read_excel("/Users/aopeng/Downloads/Dataset 3/Absenteeism_at_work.xls", usecols=['Age', 'Work load Average/day '])
```

```python
df
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
```

```python
K_col = [ x+1 for x in range(15)]
sse = []
for k in K_col:
    model = KMeans(n_clusters=k).fit(df)
    sse.append(model.inertia_)
sse = np.array(sse)
```

```python
plt.plot(K_col, sse, marker='o')
plt.xlabel('K')
plt.ylabel('SSE')
plt.show()
# k=8
```

### 2 Standardize dataset and run it with KMeans

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data_std = MinMaxScaler().fit_transform(df.values)
```

```python
model_2 = KMeans(n_clusters=8).fit(data_std)
model_2.inertia_
```

```python
plt.figure(figsize=(10,6))
plt.subplot()
plt.scatter(data_std[:, 0], data_std[:, 1], c=model_2.predict(data_std))
```

### 3 Agglomerative Clustering

```python
model_agg = AgglomerativeClustering(n_clusters=8, affinity='euclidean', linkage='single').fit_predict(data_std)
# model_agg.labels_
```

```python
plt.figure(figsize=(10,6))
plt.subplot()
plt.scatter(data_std[:, 0], data_std[:, 1], c=model_agg)
```

### 4 GaussianMixture

```python
model_gaussian = GaussianMixture(n_components=8).fit(data_std)
```

```python
pred = model_gaussian.predict(data_std)
plt.figure(figsize=(10,6))
plt.subplot()
plt.scatter(data_std[:, 0], data_std[:, 1], c=pred)
```
